Unit Testing Assignment


I created a Python calculator program and wrote unit tests using the unittest framework. I tested the add, subtract, multiply, and divide functions to ensure they returned the correct results. I executed the tests in VS Code since unittest does not run properly inside Jupyter Notebook.

What do the test results mean?

The test results show that all five unit tests passed successfully. The “Ran 5 tests” message confirms that each function was tested, and the “OK” result means the program produced the expected output for every test case. Unit testing helps identify errors early and ensures that individual parts of the code work correctly before integrating them into larger programs.
